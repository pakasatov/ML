```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics 
```

```python
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
iris.data
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.target_names
```

```python
model=KNeighborsClassifier()
model.fit(iris.data,iris.target)
```

```python
pred=model.predict(iris.data)
```

```python
pred
```

```python
metrics.confusion_matrix(iris.target,pred)
```

```python
print(metrics.classification_report(iris.target, pred))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()
G.add_nodes_from(["A","B","C","D","E"])

edges = [("A","B"),("A","C"),("B","C"),("B","D"),("C","E")]
G.add_edges_from(edges)

plt.figure(figsize=(8,6))
nx.draw(G, with_labels=True, node_color="skyblue", node_size = 1000,
        edge_color="gray", font_size=16, font_weight="bold")
plt.title("Network Graph Using Python")
plt.show()
```
