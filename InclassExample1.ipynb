```python
#example 4extra
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

data = [[2, 6],
        [4, 10],
        [6, 11],
        [8, 14],
        [10, 22],
        [12, 25]]

arr = np.array(data)
x = arr[:, 0]
y = arr[:, 1]

plt.figure(figsize=(4, 3))

plt.scatter(x, y, s=50)
plt.show()

```

```python
x = arr[:, 0].reshape(-1,1)
y = arr[:, 1].reshape(-1,1)

model = LinearRegression()
model.fit(x,y)

# จุดตัดเเกน y
intercept = model.intercept_
print(intercept)
# ค่าสัมประสิทธิ์ หรือ ความชัน
coefficient = model.coef_
print(coefficient)

# Model
print('y', '=', model.coef_,'* x', '+', model.intercept_)
```

```python
y_pred = model.coef_*x + model.intercept_
```

```python
plt.scatter(x, y, color='black')
plt.scatter(x, y_pred, color = 'blue', linewidth= 3)
plt.xlabel('feature')
plt.ylabel('target')
```
